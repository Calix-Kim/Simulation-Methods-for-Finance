```python
import numpy as np
def ELS(r,q1,q2,sigma1,sigma2,rho,M,n):
    S1_start = 100
    S2_start = 100
    principal = 100    
    
    check_term = 0.5
    check_cnt = 0
    
    case=6
    payment = np.zeros(M)
    prob = np.zeros(case)
    
    for i in range(M):
        
        S1_before = S1_start
        S2_before = S2_start
        
        check_cnt = 0
        check_barrier = False
        
        for j in range(1, n*4+1):

            # Random walk
            x1 = np.random.randn()
            x2 = np.random.randn()
    
            e1 =  x1
            e2 = rho*x1 + x2*np.sqrt(1-rho**2)
    
            S1_after = S1_before*np.exp( (r-(sigma1**2)/2)*(24/n)/48 + sigma1*e1*np.sqrt((24/n)/48))
            S2_after = S2_before*np.exp( (r-(sigma2**2)/2)*(24/n)/48 + sigma2*e2*np.sqrt((24/n)/48))
            
            # Check Barrier
            if (S1_after < S1_start * 0.6) or (S2_after < S2_start * 0.6):
                check_barrier = True
            
            # Check early repayment
            if j % n == 0:
                check_cnt+=1
                
                if (S1_after >= S1_start * (0.85 - ((check_cnt-1) * 0.05))) and (S2_after >= S2_start * (0.85 - ((check_cnt-1) * 0.05))):
                    payment[i] = principal * (1+0.0625*check_cnt) * np.exp(-r * check_cnt * check_term)
                    prob[check_cnt-1]+=1
                    break
            
            # Check Maturity repayment
            if check_cnt == 4:
                if (check_barrier == True) and ( (S1_after < S1_start * 0.7) or (S2_after < S2_start * 0.7) ):
                    payment[i] = principal * min(S1_after/S1_start, S2_after/S2_start) * np.exp(-r * check_cnt * check_term)
                    prob[check_cnt+1]+=1
                elif (check_barrier == False) and ( (S1_after < S1_start * 0.7) or (S2_after < S2_start * 0.7) ):
                    payment[i] = principal * np.exp(-r * check_cnt * check_term)
                    prob[check_cnt]+=1
                break
            
            S1_before = S1_after
            S2_before = S2_after               
    
    value = np.mean(payment)
    err = np.std(payment) / np.sqrt(M)
    prob = prob / M
    
    return value,err,prob
from datetime import datetime
r=0.03
q1=0.0
q2=0.0
sigma1=0.3
sigma2=0.4
rho=0.2
M=10000
n=8
t1=datetime.now()
value,err,prob=ELS(r,q1,q2,sigma1,sigma2,rho,M,n)
t2=datetime.now()
print('value = {:.2f}, err = {:.2f}'.format(value,err))
for i in range(6):
    print('prob[{:d}] = {:{width}.2%}'.format(i+1,prob[i],width=6))
print('Total sum of prob = {:.0%}'.format(sum(prob)))
print('Total computing time = {:f} seconds'.format((t2-t1).total_seconds()))
```
