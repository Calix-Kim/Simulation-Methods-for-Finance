```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import Symbol, integrate
```

# Acceptance-rejection Method

### ex) Beta distribution

1. Probability Density Function (We have PDF)

```python
def beta_func(alpha_1, alpha_2, x):
    x = Symbol('x')
    beta = x**(alpha_1-1) * (1-x)**(alpha_2-1)
    return integrate(beta, (x, 0, 1))
```

```python
def beta_pdf(alpha_1, alpha_2, x):
    return (1/beta_func(alpha_1, alpha_2, x) * x**(alpha_1-1) * (1-x)**(alpha_2-1))
```

```python
x_list = np.linspace(0, 1, 100)
y_list = beta_pdf(3, 2, x_list)
```

```python
plt.plot(x_list, y_list)
plt.title("Beta density function - 20204701")
plt.show()
```

    2-1. Generate X from g function ( We use uniform density function as g)

```python
X = np.random.rand(30000) # X is uniform random variable
```

    2-2. Generate U

```python
U = np.random.rand(30000)
```

    2-3. Accept X if U <= f(X) / c*g(X) (g(X) = 1 cos g is uniform)

```python
acceptance_status = (2 * U <= beta_pdf(3, 2, X)) # if c=2
accepted_X = acceptance_status * X
```

```python
accepted_X = accepted_X[accepted_X != 0]
```

```python
# Draw the approximate PDF to generate random variable
plt.hist(accepted_X, density=True, bins=100)
plt.title("Approximate Beta density function - 20204701")
plt.show()
```
