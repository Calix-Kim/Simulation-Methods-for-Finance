```python
import numpy as np
import matplotlib.pyplot as plt
```

# Inverse Transform Method

### ex) Exponential random variable

1. Cummulative Distribution Function (We have CDF)

```python
def exp_cdf(x, theta): # theta -> mean
    return (1-np.exp(-x/theta))
```

```python
x_list = np.linspace(0, 20, 100)
y_list = exp_cdf(x_list, 2)
```

```python
# exponential distribution function graph
plt.plot(x_list, y_list)
plt.show()
```

2. Input Uniform random variable into Inverse CDF

```python
def exp_cdf_inv(y, theta):
    return (-theta * np.log(1-y))
```

```python
y_list = np.random.rand(10000) # Generate 10000 uniform random variables
x_list = exp_cdf_inv(y_list, 2)
```

```python
# Draw the Approximate PDF of exponential random variable
plt.hist(x_list, density=True, bins=100)
plt.show()
```

3. Chech with Real PDF graph

```python
def exp_pdf(x, theta):
    return (np.exp(-x/theta)/theta)
```

```python
x_list = np.linspace(0, 20, 100)
y_list = exp_pdf(x_list, 2)

# exponential distribution function graph
plt.plot(x_list, y_list)
plt.show()
```
